```python
a = int(input('Digite um número de 1 a 7, e você será informado qual é o dia da semana:'))
if a == 1:
    print('Domingo')
elif a == 2:
    print('Segunda-feira')
elif a == 3:
    print('Terça-feira')
elif a == 4:
    print('Quarta-feira')
elif a == 5:
    print('Quinta-feira')
elif a == 6:
    print('Sexta-feira')
elif a == 7:
    print('Sábado')
else:
    print('Erro, digite um número entre 1 e 7')
```

```python
horas_trabalhadas = float(input("Digite a quantidade de horas trabalhadas: "))
valor_por_hora = float(input("Digite o valor recebido por hora: "))
if horas_trabalhadas <= 40:
    salario = horas_trabalhadas * valor_por_hora
else:
    horas_extras = horas_trabalhadas - 40
    salario = (40 * valor_por_hora) + (horas_extras * valor_por_hora * 1.5)
print(f"Salário total: R$ {salario:.2f}")
```
